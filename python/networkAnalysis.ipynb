The data simulated in the code below represents a fictional social network where individuals are nodes, and friendships are represented as edges between these nodes. Each person (node) has an identifier (id) and a name (name). The relationships (edges) between individuals indicate friendships.

```python
pessoas = [
{ "id": 0, "name": "Maria" },
{ "id": 1, "name": "Carlão" },
{ "id": 2, "name": "Vitinho" },
{ "id": 3, "name": "Isabela" },
{ "id": 4, "name": "Thales" },
{ "id": 5, "name": "Clara" },
{ "id": 6, "name": "Henrique" },
{ "id": 7, "name": "Duda" },
{ "id": 8, "name": "Kate" },
{ "id": 9, "name": "Kleitin" }
]


amizade = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4),
(4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]
```

```python
import networkx as nx
import matplotlib.pyplot as plt


# Criando um grafo não direcionado
G = nx.Graph()

# Adicionando nós com atributos de nome
for user in pessoas:
    G.add_node(user["id"], name=user["name"])

# Adicionando arestas
G.add_edges_from(amizade)




# Desenhar o grafo
pos = nx.spring_layout(G)  # Define a posição dos nós (layout)

# Desenhar nós
nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=500)

# Desenhar arestas
nx.draw_networkx_edges(G, pos, edgelist=amizade, edge_color='gray')

# Adicionar etiquetas aos nós
labels = nx.get_node_attributes(G, 'name')
nx.draw_networkx_labels(G, pos, labels, font_size=12)

# Mostrar o grafo
plt.style.use('ggplot')
plt.title('Grafo de Amizades entre Usuários')
plt.show()
```

Exemplo de análise que o grafo nos permite fazer: note que, pela imagem anterior, para que a Maria vire amiga do Henrique é necessário que a mesma conheça, antes, o Vitinho, a Isabela, o Thales e a Clara.
