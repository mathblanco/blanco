#Grafos

Grafos são estruturas matemáticas que consistem em nós (ou vértices) conectados por arestas. Essa representação é fundamental para modelar e analisar uma ampla gama de problemas complexos em diversas áreas, como redes sociais, logística, biologia molecular, telecomunicações e muito mais. As aplicações dos grafos são vastas: eles são usados para otimizar rotas em mapas, recomendar amigos em redes sociais, projetar redes de computadores eficientes, prever interações entre moléculas em biologia, entre outros. A flexibilidade dos grafos em capturar relações e interações torna-os uma poderosa ferramenta para resolver problemas que envolvem a organização e a conexão de elementos.

Exemplo:
Os dados simulados no código abaixo representam uma rede social fictícia onde os indivíduos são nós, e as amizades são representadas como arestas entre esses nós. Cada pessoa (nó) possui um identificador (id) e um nome (name). As relações (arestas) entre os indivíduos indicam amizades.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Definindo os dados
pessoas = [
    {"id": 0, "name": "Maria"},
    {"id": 1, "name": "Carlão"},
    {"id": 2, "name": "Vitinho"},
    {"id": 3, "name": "Isabela"},
    {"id": 4, "name": "Thales"},
    {"id": 5, "name": "Clara"},
    {"id": 6, "name": "Henrique"},
    {"id": 7, "name": "Duda"},
    {"id": 8, "name": "Kate"},
    {"id": 9, "name": "Kleitin"}
]

amizade = [
    (0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4),
    (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)
]

# Criando um grafo não direcionado
G = nx.Graph()

# Adicionando nós com atributos de nome
for user in pessoas:
    G.add_node(user["id"], name=user["name"])

# Adicionando arestas
G.add_edges_from(amizade)

# Identificando os grupos
grupo1 = [0, 1, 2, 3]  # Maria, Carlão, Vitinho, Isabela
grupo2 = [5, 6, 7, 8, 9]  # Clara, Henrique, Duda, Kate, Kleitin

# Definindo cores para os nós de cada grupo
node_colors = ['skyblue' if node in grupo1 else 'lightcoral' if node in grupo2 else 'gray' for node in G.nodes()]

# Desenhar o grafo
pos = nx.spring_layout(G)  # Define a posição dos nós (layout)

# Desenhar nós
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=500)

# Desenhar arestas
nx.draw_networkx_edges(G, pos, edgelist=amizade, edge_color='gray')

# Adicionar etiquetas aos nós
labels = nx.get_node_attributes(G, 'name')
nx.draw_networkx_labels(G, pos, labels, font_size=12)

# Mostrar o grafo
plt.style.use('ggplot')
plt.title('Grafo de Amizades entre Usuários')
plt.show()
```

Exemplo de análise que o grafo nos permite fazer: note que, pela imagem anterior, para que a Maria vire amiga do Henrique é necessário que a mesma conheça, antes, o Vitinho, a Isabela, o Thales e a Clara.
Note também que no gráfico apresentado, as cores distintas destacam dois grupos principais de indivíduos com base nas relações de amizade: o grupo azul (Maria, Carlão, Vitinho, Isabela) e o grupo vermelho (Clara, Henrique, Duda, Kate, Kleitin). Cada cor representa uma comunidade coesa dentro da rede social representada pelo grafo, onde os nós coloridos em azul claro e vermelho claro indicam forte interconexão interna entre seus membros. Os nós cinzas representam indivíduos que têm conexões na rede, mas não estão diretamente ligados aos grupos identificados, revelando a estrutura de subgrupos e conexões na rede de amizades.

Outro exemplo: Vendas

```python
# Definindo os dados dos produtos e vendedores
produtos = [
    {"id": 0, "name": "Smartphone"},
    {"id": 1, "name": "Tablet"},
    {"id": 2, "name": "Notebook"},
    {"id": 3, "name": "Câmera"},
    {"id": 4, "name": "Impressora"},
    {"id": 5, "name": "Fone de Ouvido"},
    {"id": 6, "name": "Mouse"},
    {"id": 7, "name": "Teclado"},
    {"id": 8, "name": "Monitor"},
    {"id": 9, "name": "Carregador"}
]

vendas = [
    (0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4),
    (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)
]

# Criando um grafo não direcionado
G = nx.Graph()

# Adicionando nós com atributos de nome
for produto in produtos:
    G.add_node(produto["id"], name=produto["name"])

# Adicionando arestas representando vendas
G.add_edges_from(vendas)

# Identificando os grupos de produtos
grupo1 = [0, 1, 2, 3]  # Smartphone, Tablet, Notebook, Câmera
grupo2 = [5, 6, 7, 8, 9]  # Fone de Ouvido, Mouse, Teclado, Monitor, Carregador

# Definindo cores para os nós de cada grupo
node_colors = ['skyblue' if node in grupo1 else 'lightcoral' if node in grupo2 else 'gray' for node in G.nodes()]

# Desenhar o grafo
pos = nx.spring_layout(G)  # Define a posição dos nós (layout)

# Desenhar nós
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=500)

# Desenhar arestas
nx.draw_networkx_edges(G, pos, edgelist=vendas, edge_color='gray')

# Adicionar etiquetas aos nós
labels = nx.get_node_attributes(G, 'name')
nx.draw_networkx_labels(G, pos, labels, font_size=9)

# Mostrar o grafo
plt.style.use('ggplot')
plt.title('Grafo de Vendas de Produtos')
plt.show()
```

A análise dos grupos identificados no grafo de vendas de produtos pode ser usada para recomendar produtos de forma estratégica, especialmente ao entender quais produtos têm associações mais fortes entre si. Ao identificar grupos como o grupo azul (Smartphone, Tablet, Notebook, Câmera) e o grupo vermelho (Fone de Ouvido, Mouse, Teclado, Monitor, Carregador), é possível inferir padrões de compra que podem ser explorados em estratégias de recomendação.

Por exemplo, se um cliente comprar um Smartphone (do grupo azul), a análise sugere que ele também poderia se interessar por produtos como Tablet ou Notebook, pois esses produtos estão frequentemente associados em vendas. Da mesma forma, se um cliente adquire um Teclado (do grupo vermelho), a recomendação poderia incluir produtos como Fone de Ouvido ou Monitor, aproveitando a relação de vendas entre esses itens.

Essa análise ajuda a personalizar recomendações de produtos tanto em ambientes físicos quanto online, melhorando a experiência do cliente ao oferecer sugestões relevantes e aumentando a eficiência das estratégias de marketing ao identificar produtos complementares que podem aumentar o valor da cesta de compras do cliente.
